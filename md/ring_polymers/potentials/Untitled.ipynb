```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
lammps11 = np.loadtxt('lammps_pot11.dat',skiprows = 6)[:,1:]
lammps12 = np.loadtxt('lammps_pot12.dat',skiprows = 6)[:,1:]
lammps22 = np.loadtxt('lammps_pot22.dat',skiprows = 6)[:,1:]
cg11 = np.loadtxt('vEff_M1-50_M2-50_K1-00_K2-00_fit.dat',skiprows = 3)[:,1:]
cg12 = np.loadtxt('vEff_M1-50_M2-50_K1-00_K2-30_fit.dat',skiprows = 3)[:,1:]
cg22 = np.loadtxt('vEff_M1-50_M2-50_K1-30_K2-30_fit.dat',skiprows = 3)[:,1:]
lammps = [lammps11,lammps12,lammps22]
cg = [cg11,cg12,cg22]
```

```python
fig,ax = plt.subplots(1,2, figsize = (20,10))
for ind,types in enumerate(['11','12','22']):
    ax[0].plot(lammps[ind][:,0],lammps[ind][:,1], label = f'lammps{types}',linewidth = 1.0)
    ax[0].plot(cg[ind][:,0],cg[ind][:,1], 'x', markersize = 0.5,linestyle = 'None',label = f'cg{types}')
    ax[0].set_xlabel('r[sig]')
    ax[0].set_ylabel('Potential')
    ax[0].legend()
for ind,types in enumerate(['11','12','22']):
    ax[1].plot(lammps[ind][:,0],lammps[ind][:,2], label = f'lammps{types}', linewidth = 1.0)
    ax[1].plot(cg[ind][:,0],cg[ind][:,2], marker = 'x',linestyle = 'none',markersize = 0.5,label = f'cg{types}')
    ax[1].set_xlabel('r[sig]')
    ax[1].set_ylabel('Force')
    ax[1].set_ylim(-2,5)
    ax[1].legend()
plt.savefig('lammps.pdf')
    
```
