```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
plt.style.use('bmh')
plt.rcParams.update({
    "text.usetex": True,
    "font.family": "sans-serif",
    "font.size": 20,
    "axes.titlesize": 20,
    "axes.labelsize": 20,
    "font.sans-serif": ["Helvetica"],
    "axes.facecolor": '#ffffff',
    "figure.autolayout": True,
	})
```

```python
def return_name(N,L,types):
    return 'SP_N{:},{:}_L{:}_f166,200_Sqs_{:}.dat'.format(N[0],N[1],L,types)
```

```python
N = [400,112]
Ls = [12.38,11.35]
```

```python
for L in Ls:
    data = np.loadtxt('./data/' + return_name(N,L,1))
    plt.errorbar(data[:,0],data[:,1],yerr=data[:,2],label = '$\\rho = {:.2f}$'.format(np.sum(N)/L**3))
    plt.ylim(0,5)
    plt.xlabel('q')
    plt.ylabel('S(q)')
plt.legend()
plt.savefig('Sq.pdf')
```
